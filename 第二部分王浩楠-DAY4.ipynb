为什么随着batch的增加，训练时会更加稳定？
个人想法：因为随着batch的增加，每次的训练都会成为训练好在进行测试，其每次被喂数据而得到的拟合函数所得的各图像得分是符合更多图片的；而batch小时就如同一个没有训练到位的学生一般，其总结的函数只适用于一些特例，当再喂一些数据后就会发现他的函数需要改变，改变后就会出现比较大的波动
